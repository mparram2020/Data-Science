```python
# read_html allows to scrape data
tablas = pd.read_html('https://web.archive.org/web/20240605004535/https://en.wikipedia.org/wiki/UEFA_Euro_2024')
```

```python
# Find all the groups
tablas[19]
tablas[26]
tablas[33]
tablas[40]
tablas[47]
tablas[54]
```

```python
tablas[19]
```

```python
# Get all the groups in the tournament
# Rename column Puntos as Pts
# Create a dictionary to insert the groups
dict_tablas = {}
for letra, i in zip(alfabeto, range(19, 60, 7)):
  df = tablas[i]
  df.rename(columns={df.columns[1]: 'Team'}, inplace=True)
  dict_tablas[f'Group {letra}'] = df
  print(df,'\n')
```

```python
tablas[19]
```

```python
with open('dict_table', 'wb') as output:
  pickle.dump(dict_tablas, output)
```
